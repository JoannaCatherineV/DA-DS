```python
#MAIN MAIN MAIN MAIN
class Flights:
  def __init__(self, number, departure, arrival, date, time, seats):
    self.number=number
    self.departure=departure
    self.arrival=arrival
    self.date=date
    self.time=time
    self.seats=seats
    self.booked_seats=0

  def __repr__(self):
    return f"Flight {self.number}: {self.departure} to {self.arrival} on {self.date} at {self.time} ({self.seats} seats)\n"

class Passenger:
  def __init__(self):
    self.name = input("Enter your Name:")
    self.email = input("Enter your Email Id:")
    self.phone = int(input("Enter your Phone Number:"))

class Reservation:
  def __init__(self):
    self.flights=[]
    self.passengers={}
    self.passenger_seats = {}
  #add flight
  def add_flight(self,flight):
    self.flights.append(flight)
  #add passenger
  def add_passenger(self,passenger):
    self.passengers[passenger.name]={"Email": passenger.email, "Phone": passenger.phone}

  #book flight
  def book_flight(self, passenger):
    print("\n")
    print("Book your Flight! Here is the list of Departure and Arrival Cities:")
    for i,j in enumerate(self.flights):
      print(f"{i+1}.{j.departure} to {j.arrival}")
    print("\n")

    while True:
      departure = input("Enter your departure city: ")
      arrival = input("Enter your destination city: ")
      find_flight = [i for i in self.flights if i.departure == departure and i.arrival == arrival]

      if find_flight:
        print("\n")
        print("Available flights for your entry:")
        for i, j in enumerate(find_flight):
          print(f"{i+1}. Flight {j.number}: {j.departure} to {j.arrival} on {j.date} at {j.time}, {j.seats - j.booked_seats} seats available)")
        available_flight = [flight.number for flight in find_flight]
        break
      else:
        print("Please enter a valid location from the list of available Departure and Arrival cities.\n")
    print("\n")
    while True:
      choose_flight=input("Please enter your preferred flight number:")
      if choose_flight in available_flight:
        choice = next(flight for flight in find_flight if flight.number == choose_flight)
        if choice:
          booked_seats = [i['Seat Number'] for i in self.passenger_seats.values() if i['Flight Number']==choice.number]
          print("Seats : 1(A-F) to 25(A-F)")
          if booked_seats:
            print("Booked seats:")
            for i in booked_seats:
              print(i)
        if choice:
          if choice.booked_seats < choice.seats:
            seat_number = input("Enter your preferred seat number: ")
            aa=[booking for passenger, booking in self.passenger_seats.items() if booking['Flight Number'] == choice.number and booking['Seat Number'] == seat_number]
            if aa:
              print(f"Seat {seat_number} is already booked. Please enter a different seat number or Different flight.")
              continue
            else:
              choice.booked_seats += 1
              self.passenger_seats[passenger.name] = {'Seat Number': seat_number, 'Flight Number': choice.number}

              print(f"\nFlight {choice.number} booked for {passenger.name} with seat number {seat_number}!")
            break
        else:
          print("No seats available")
          break
      else:
        print("Invalid flight number. Please enter from the list of Available flights !!")

  def view_booking(self,passenger):
    print("\n")
    print("*** View your Booking ***")
    if passenger.name in self.passenger_seats:
      flight_number = self.passenger_seats[passenger.name]['Flight Number']
      flight = next((f for f in self.flights if f.number == flight_number), None)

      if flight:
        seat_number = self.passenger_seats[passenger.name]['Seat Number']
        print(f"Flight: {flight.number}\nPassenger: {passenger.name}\nDeparture: {flight.departure}\nArrival: {flight.arrival}\nDate: {flight.date}\nTime: {flight.time}\nSeat Number: {seat_number}")
      else:
        print("Flight not found")
    else:
      print("No booking found for this passenger")



reserve=Reservation()
f1=Flights("AA101", "New York", "Los Angeles", "2024-09-01", "10:00", 150)
f2=Flights("AA102", "Chennai", "Mumbai", "2024-09-01", "14:00", 150)
f3=Flights("AA103", "Goa", "Hyderabad", "2024-09-02", "12:00", 150)
f4=Flights("AA104", "New York", "Los Angeles", "2024-09-01", "16:00", 150)

reserve.add_flight(f1)
reserve.add_flight(f2)
reserve.add_flight(f3)
reserve.add_flight(f4)

p1=Passenger()
reserve.add_passenger(p1)
reserve.book_flight(p1)
reserve.view_booking(p1)
print("\n")

```
